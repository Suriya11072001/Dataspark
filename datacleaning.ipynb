```python
import pandas as pd
```

```python
sales=pd.read_csv('Sales.csv')
products=pd.read_csv('Products.csv')
exchange_rates=pd.read_csv('Exchange_Rates.csv')
stores=pd.read_csv('Stores.csv')
customers=pd.read_csv('Customers.csv',encoding='ISO-8859-1')
```

```python
sales.isnull().sum()
```

```python
sales.dtypes
```

```python
sales['Order Date'] = pd.to_datetime(sales['Order Date'], format='%m/%d/%Y')
sales['Delivery Date']=pd.to_datetime(sales['Delivery Date'],format='%m/%d/%Y',errors='coerce')
sales['Currency Code']=sales['Currency Code'].astype(str)
```

```python
sales.dtypes
```

```python
sales.fillna('Not available',inplace=True)
```

```python
sales.isnull().sum()
```

```python
sales
```

```python
products.info()
```

```python
products['Unit Cost USD']=products['Unit Cost USD'].replace('[\$,]', '', regex=True).astype(float)
products['Unit Price USD']=products['Unit Price USD'].replace('[\$,]', '', regex=True).astype(float)
```

```python
products.info()
```

```python
products.isnull().sum()
```

```python
products
```

```python
stores.isnull().sum()
```

```python
stores.fillna(0,inplace=True)
```

```python
stores.dtypes
```

```python
stores['Open Date']=pd.to_datetime(stores['Open Date'],format='%m/%d/%Y')
```

```python
stores.dtypes
```

```python
stores
```

```python
exchange_rates.isnull().sum()
```

```python
exchange_rates.dtypes
```

```python
exchange_rates['Date']=pd.to_datetime(exchange_rates['Date'],format='%m/%d/%Y')
```

```python
exchange_rates.dtypes
```

```python
exchange_rates
```

```python
customers.isnull().sum()
```

```python
customers['State Code'].fillna('Not available',inplace=True)
```

```python
customers.dtypes
```

```python
customers['Birthday']=pd.to_datetime(customers['Birthday'],format='%m/%d/%Y')
```

```python
customers.info()
```

```python
customers
```

```python
overall=sales
```

```python
overall=pd.merge(overall,customers,on='CustomerKey',how='inner')
```

```python
overall
```

```python

overall=pd.merge(overall,exchange_rates,left_on=['Order Date','Currency Code'],right_on=['Date','Currency'],how='inner')
```

```python
overall
```

```python
overall=pd.merge(overall,products,on='ProductKey',how='inner')
```

```python
overall
```

```python
overall=pd.merge(overall,stores,on=['StoreKey','Country','State'],how='inner')
```

```python
overall
```

```python
overall.columns
```

```python
overall.dtypes
```

```python
col=list(overall.columns)
col1=[]
for c in col:
    col1.append(c.replace(' ','_').lower())
    
overall.columns=col1
```

```python
col=list(stores.columns)
col1=[]
for c in col:
    col1.append(c.replace(' ','_').lower())
    
stores.columns=col1
```
